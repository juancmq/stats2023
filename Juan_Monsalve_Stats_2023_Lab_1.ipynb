```python
plt.plot(x,y)
plt.title("Graph of x**2"); #this is to label the graph
plt.xlabel("x"); # this is to label the x-axis
plt.ylabel("y"); # this is to label the y-axis
```

```python
plt.plot(x,y,label="x squared") # use this to label what the line means
plt.title("Graph of x**2");
plt.xlabel("x");
plt.ylabel("y");
plt.legend() # use this for the label of the lines to appear
```

```python
x = np.linspace(0, 2, 100) # series of numbers

plt.plot(x, x, label='linear') #you can also label individual lines
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**8, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend() # always use this for the different meanings of the line
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^'); # you can also change each line's shape and color with this order
```

```python
names = ['cat', 'dog', 'bird'] # categories in the graph
values = [15, 18, 6]

plt.figure(1, figsize=(9, 3)) # this is for how big the image will be


plt.subplot(131)
plt.bar(names, values) # use this to make a bar graph

plt.subplot(132)
plt.scatter(names, values) # use this to make a scatter plot graph

plt.subplot(133)
plt.plot(names, values) # use this to make a bent line


plt.suptitle('Categorical Plotting')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2*np.pi*t)

plt.plot(t, s)
plt.title(r'$\alpha_i > \beta_i$', fontsize=20)
plt.text(1, -0.6, r'$\sum_{i=0}^\infty x_i$', fontsize=20)
plt.text(0.5, 0.6, r'$\mathcal{A}\mathrm{sin}(2 \omega t)$',fontsize=15)
plt.xlabel('time (s)')
plt.ylabel('volts (mV)')
plt.show()
```
