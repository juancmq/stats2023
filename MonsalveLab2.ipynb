```python
import matplotlib.pyplot as plt

labels =['Frogs', 'Dogs', 'Cats', 'Birds']

sizes = [15,30,45,10]

explode=(0,0.1,0,0)

figl,axl = plt.subplots ()
axl.pie(sizes,
        explode=explode,
        labels=labels,
        autopct='%1.1f%%',
        shadow=True, 
        startangle=90)
axl.axis('equal')
```

```python
import matplotlib.pyplot as plt

labels =['Barcelona', 'Madrid', 'Bilbao', 'Seville'] # Names for you pie pieces

sizes = [50,30,5,15] # How much each percentage is

explode=(0.05,0,0,0) # How far each piece is from one another

figl,axl = plt.subplots ()
axl.pie(sizes,
        explode=explode,
        labels=labels,
        autopct='%1.1f%%',
        shadow=True, 
        startangle=90)
axl.axis('equal')
```
